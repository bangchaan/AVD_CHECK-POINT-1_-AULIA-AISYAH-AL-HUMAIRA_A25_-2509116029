# **Business Understanding**

## **Business Objective**

Dataset ini memuat informasi mengenai beberapa brand smartphone seperti Samsung, Xiaomi, Realme, Apple, Oppo, Vivo, dan lainnya. Terdapat komponen seperti RAM, storage, nama model, dan harga.

Analisis ini dibuat dengan tujuan untuk mengklasifikasikan data-data dari brand *smartphone*. Data yang terkumpul diolah menjadi sebuah informasi yang dapat membantu dalam memprediksi harga, analisis pasar, ataupun perbandingan harga *smartphone* lainnya dengan cara mengidentifikasi tiap komponen apa yang dapat mempengaruhi sebuah harga *smartphone*

## **Assess Situation**

 Dataset yang digunakan berasal dari satu sumber website Perancis, sehingga harga yang tersedia tidak sepenuhnya merepresentasikan harga pasar global. Selain itu, dataset tidak memuat beberapa variabel penting yang secara teori mempengaruhi harga smartphone seperti prosesor, kamera, kapasitas baterai, dan tahun rilis.

Terdapat juga variabel seperti warna yang kemungkinan tidak memiliki pengaruh signifikan terhadap harga. Selain itu, adanya brand premium seperti Apple berpotensi menjadi outlier yang dapat mempengaruhi hasil analisis. Karena keterbatasan tersebut, analisis akan difokuskan pada variabel yang tersedia yaitu RAM, storage, dan brand sebagai faktor yang diasumsikan mempengaruhi harga *smartphone*.
