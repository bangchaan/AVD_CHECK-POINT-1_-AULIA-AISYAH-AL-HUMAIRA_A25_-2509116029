### **Outliers Values**

Pada tahap ini dilakukan identifikasi nilai ekstrem (outliers) pada kolom yang bertipe data numerik, seperti RAM, Storage, dan Final Price. Pemeriksaan ini bertujuan untuk insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.


Akan tetapi,  dalam industri smartphone, variasi spesifikasi dan harga yang cukup besar merupakan hal yang wajar, terutama karena adanya perbedaan segmen pasar antar level sehingga tidak perlu ditangani.

Berdasarkan hasil identifikasi sebagai berikut:

* RAM: 0%
* Storage: sekitar 0,9%
* Final Price: sekitar 5,1% (kurang dari 1%)

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# **Eksplorasi Data (EDA)**

## **Comparison/Perbandingan**

Aktivitas:
Membandingkan rata-rata harga (Final Price) berdasarkan brand smartphone.

Tujuan:
Mengidentifikasi brand mana yang memiliki harga rata-rata tertinggi serta membandingkan segmentasi harga antar brand secara langsung.

Visualisasi:
Bar Chart (Grafik Batang).

```python
harga_brand = df.groupby('Brand')['Final Price'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=harga_brand.index, y=harga_brand.values, palette='viridis', hue=harga_brand.index, legend=False)
plt.title('Harga Smartphone berdasarkan Brand')
plt.xlabel('Brand')
plt.ylabel('Final Price')
plt.xticks(rotation=45)
plt.show()
```


Berdasarkan grafik perbandingan rata-rata harga, beberapa brand seperti Apple dan Lenovo menempati posisi teratas dengan nilai rata-rata yang lebih tinggi dibandingkan brand lainnya. Hal ini menunjukkan bahwa kedua brand tersebut berfokus pada segmen flagship atau katagori high-end.

Sementara brand seperti Alcatel, Qubo, dan SPC memiliki rata-rata harga yang cenderung lebih rendah. Hal ini menujukkan segmen pasar brand-brand tersebut berada di entry-level atau terjangkau.

Dari data tersebut, dapat diketahui bahwa terdapat perbedaan segmentasi pasar di industri smartphone. Masing-masing brand memiliki strategi harga yang berbeda, disesuaikan dengan target pasar yang ingin mereka capai.

## **Composition/Komposisi**


Aktivitas: Melihat proporsi atau persentase harga akhir berdasarkan brand smartphone.

Tujuan: Mengetahui brand mana yang memiliki kontribusi harga terbesar di dataset.

Visualisasi: Pie Chart (Grafik Lingkaran).

```python
price_by_brand = df.groupby('Brand')['Final Price'].mean().sort_values(ascending=False)
price_by_brand_top3 = price_by_brand.head(3)

plt.figure(figsize=(10, 8))
price_by_brand_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi 3 Kategori Brand Teratas berdasarkan Harga')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

Berdasarkan pie chart, dapat diketahui bahwa proporsi rata-rata harga dari tiga brand teratas cukup seimbang. Apple memiliki porsi sedikit lebih besar dibandingkan Lenovo dan Asus, namun selisihnya tidak terlalu jauh.

Hal ini menunjukkan bahwa ketiga brand tersebut berada pada rentang harga yang relatif sama, dengan Apple sedikit lebih tinggi dalam rata-rata harga.
